```python
import pandas as pd
transaction_df = pd.read_csv('TokenAnalystData.csv', index_col=0, converters={'value':float})
```

```python
address_df = transaction_df.groupby('from_address')['value'].agg(['min', 'max', 'mean'])
address_df.columns = ['min_value', 'max_value', 'mean_value']
address_df.head()
```

```python
address_df['total_transactions'] = transaction_df['from_address'].value_counts()
```

```python
address_df.head()
```

```python
address_df[['mean_trans_per_block', 'max_trans_per_block']] = transaction_df.groupby(['from_address', 'block_number']).size().reset_index(name='counts').groupby('from_address')['counts'].agg(['mean', 'max'])
address_df['num_blocks'] = transaction_df.groupby('from_address')['block_number'].nunique()
address_df['num_recipients'] = transaction_df.groupby('from_address')['to_address'].nunique()
address_df[['min_gas', 'max_gas', 'mean_gas']] = transaction_df.groupby('from_address')['gas'].agg(['min', 'max', 'mean'])
address_df[['min_gas_price', 'max_gas_price', 'mean_gas_price']] = transaction_df.groupby('from_address')['gas_price'].agg(['min', 'max', 'mean'])
```
