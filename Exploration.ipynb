```python
transaction_df = pd.read_csv('TokenAnalystData.csv', index_col=0, converters={'value':float})
transaction_df.head(40)
```

```python
transaction_df.dtypes
```

```python
# Value Data
transaction_df.groupby('from_address')['value'].agg(['min', 'max', 'mean'])
```

```python
# Number of transactions (in 27 mins...some of these values or way too big for humans,
# anything over 10 seems unrealistic for sure)
transaction_df['from_address'].value_counts()
```

```python
print('Num Blocks: ', transaction_df['block_number'].nunique())
print('Num Addresses: ', transaction_df['from_address'].nunique())
```

```python
# transactions per block (mean and max)
# again there's some very large numbers here, when would a human be making more than 1 transaction per block?
transaction_df.groupby(['from_address', 'block_number']).size().reset_index(name='counts').groupby('from_address')['counts'].agg(['mean', 'max']).sort_values('max', ascending=False)
```

```python
# number of blocks address appears in
# measuring how consistently transactions are created
transaction_df.groupby('from_address')['block_number'].nunique().sort_values(ascending=False)
```

```python
# clearly no human is creating transactions to over 500 unique addresses
transaction_df.groupby('from_address')['to_address'].nunique().sort_values(ascending=False)
```
